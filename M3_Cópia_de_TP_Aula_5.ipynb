#Utilizando Apply para aplicar funções em Data Frames

```python
def verifica_aprovacao(a):
  if(a >= 75):
    return 'muito bom'
  elif(a >= 50 ):
    return 'mediano';
  else:
    return 'ruim'
```

```python
def verifica_ano_filme(a):
  if(a >= 2020):
    return 'novo'
  elif(a >= 2010):
    return 'mais ou menos novo'
  else:
    return 'antigo'
```

```python
df['aprovacao do publico'] = df['Rotten Tomatoes %'].apply(verifica_aprovacao)
```

```python
df['analise da idade'] = df['Year'].apply(verifica_ano_filme)
```

```python
df.head(10)
```

```python
df.to_csv('sample_data/movies_novo.csv', encoding='UTF-8', sep=',')
```

#Merge

```python
# unindo dois dataframes
df_animal1 = pd.DataFrame([['gato', 4], 
                           ['cachorro', 4], 
                           ['pato', 2]],
                          columns=['animal', 'patas'])
df_animal2 = pd.DataFrame([['gato', 'mamífero'], 
                           ['cachorro', 'mamífero'], 
                           ['pato', 'ave'],
                           ['cavalo', 'mamífero']],
                          columns=['animal', 'tipo'])
```

```python
df_animal1
```

```python
df_animal2
```

```python
# especificando as colunas da união
df_animal1.merge(df_animal2, left_on='animal', right_on='animal')
```

```python
# unindo os dados com chave semelhante
df_animal1.merge(df_animal2, how='inner', on='animal')
```

```python
# adicionando mais uma linha na tabela 1
df_animal3 = pd.DataFrame([['cavalo', 4], 
                           ['ganso', 2]],
                          columns=['animal', 'patas'])
df_animal3
```

```python
# juntando os novos dados 
df_animal1 = pd.concat([df_animal1, df_animal3], ignore_index=True)
df_animal1
```

```python
df_animal2
```

```python
# unindo os dados com chave semelhante
df_animal1.merge(df_animal2, how='inner', on='animal')
```

```python
# unindo os dados com chave semelhante, modificando a forma da união
df_animal1.merge(df_animal2, how='left', on='animal')
```

#Pivot

```python
# alterando a exibição de linhas para colunas
df_letra_numero = pd.DataFrame({'abc': ['abc','abc','abc','abc','abc','abc'],
                                'letras': ['A', 'B', 'C', 'D', 'E', 'F'],
                                'numeros': [1, 2, 3, 4, 5, 6]})
df_letra_numero
```

```python
# transformando linhas em colunas
df_letra_numero.pivot(index='abc', columns='letras', values='numeros')
```

```python
# alterando a exibição de linhas para colunas
df_vendas = pd.DataFrame({'mes': ['01-jan','02-fev','02-fev','04-abr','04-abr','05-mai'],
                          'produto': ['celular', 'celular', 'notebook', 'notebook', 'mouse', 'celular'],
                          'quantidade': [10, 25, 33, 14, 5, 16]})
df_vendas
```

```python
df_vendas.pivot(index='produto', columns='mes', values='quantidade')
```

#Melt

```python
# alterando a exibição de colunas para linhas
df_temperatura = pd.DataFrame([[20,21,24,26,22,19], 
                               [29,33,34,36,29,25]],
                              columns=['jan', 'fev', 'mar', 'abr', 'mai', 'jun'],
                              index=['min', 'max'])
df_temperatura
```

```python
# transformando colunas em linhas
df_temperatura.melt(value_vars=['jan', 'fev', 'mar', 'abr', 'mai', 'jun'])
```

```python
# transformando colunas em linhas, incluindo index
df_temperatura.melt(value_vars=['jan', 'fev', 'mar', 'abr', 'mai', 'jun'], ignore_index=False)
```

```python
# transformando colunas em linhas, incluindo index no dataframe
df_temperatura.melt(value_vars=['jan', 'fev', 'mar', 'abr', 'mai', 'jun'], ignore_index=False).reset_index()
```

```python
# transformando colunas em linhas, ajustando colunas do dataframe
df_temperatura\
  .melt(value_vars=['jan', 'fev', 'mar', 'abr', 'mai', 'jun'], 
        ignore_index=False)\
  .reset_index()\
  .rename(columns={'index':'min/max','variable':'mes','value':'temperatura'})
```

### Estudo de Caso

Com um conjunto de dados, como incluir uma nova observação, realizar merge com outra tabela contendo informações extras a respeito dessa observação?
*   https://www.kaggle.com/datasets/sveta151/tiktok-popular-songs-2022
*   https://www.kaggle.com/datasets/sveta151/tiktok-popular-songs-2021



```python
tt21 = pd.read_csv('sample_data/TikTok_songs_2021.csv')
tt21.head()
```

```python
tt21['year'] = 2021
tt21.head()
```

```python
tt21 = tt21[['track_name', 'artist_name', 'year']]
tt21.head()
```

```python
tt22 = pd.read_csv('sample_data/TikTok_songs_2022.csv')
tt22.head()
```

```python
tt22['year'] = 2022
tt22
```

```python
tt22 = tt22[['track_name', 'artist_name', 'year']]
tt22.head()
```

```python
tt22.merge(tt21, how='left', on=['track_name','artist_name']).rename(columns={'year_x':'ano_1','year_y':'ano_2'})
```

```python
novo_arquivo = tt22.merge(tt21, how='inner', on=['track_name','artist_name']).rename(columns={'year_x':'ano_1','year_y':'ano_2'})
```

```python
novo_arquivo
```

```python
novo_arquivo.to_csv('sample_data/musicas_que_bombaram.csv', encoding='UTF-8', sep=',')
```

```python
populares_brasil = netflix_most_popular.merge(populares_brasil, how="inner", on='show_title')
```
