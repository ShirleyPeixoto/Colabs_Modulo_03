### Importar biblioteca


```python
import pandas as pd
import numpy as np
```

### Crie 3 séries diferentes, cada uma com comprimento 100, como segue:
1. O primeiro com números aleatórios de 1 a 4
2. O segundo com números aleatórios de 1 a 3
3. O terceiro com número aleatórios de 10.000 a 30.000

```python
s1 = np.random.randint(1,5,100)
s1
```

```python
s2 = np.random.randint(1,4,100)
s2
```

```python
s3 = np.random.randint(10000,30001,100)
s3
```

### Vamos criar um DataFrame juntando as Séries por coluna

```python
df = pd.DataFrame(data = {'s1':s1, 's2':s2 , 's3':s3})
df
```

### Altere o nome das colunas para bedrs, bathrs, price_sqr_meter

```python
df = df.rename(columns = {'s1':'bedrs'})
df = df.rename(columns = {'s2':'bathrs'})
df = df.rename(columns = {'s3':'price_sqr_meter'})
df
```

### Crie um DataFrame de uma coluna com os valores das 3 Series e o chame de `GrandeColuna`

```python
df2 = pd.DataFrame()
df2['GrandeColuna']=df['bedrs']+df['bathrs']+df['price_sqr_meter']
df2
```
