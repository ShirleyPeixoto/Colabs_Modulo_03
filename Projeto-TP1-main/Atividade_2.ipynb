### Crie uma função chamada `Maioridade` que retorne um valor booleano para uma nova coluna chamada `legal_drinker` (Considere maioridade como maior ou igual a 18 anos)

```python
def Maioridade (age):
  if age >= 18:
    return True
  else:
    return False

df['legal_drinker'] = df['age'].apply(Maioridade)
df.head(4)
```
