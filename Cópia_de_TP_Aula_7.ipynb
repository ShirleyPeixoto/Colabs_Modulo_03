```python
df.fillna(-1).head(20)
```

#Explorando ainda mais a utilização das funções acima

Caso queiramos alterar elementos de um `DataFrame` (seja uma tabela, uma lista de elementos, ou uma única célula), devemos usar um dentre `at`, `iat`, `loc` e `iloc`.

#MAP

Usado para substituir cada valor em uma série por outro valor, que pode ser derivado de uma função, um `dict` ou uma `série`.

```python
import numpy as np
```

```python
serie = pd.Series(['cachorro', 'gato', np.nan, 'coelho','cavalo'])
serie
```

```python
serie.map('Eu tenho um {}'.format)
```

```python
serie.map('Eu tenho um {}'.format, na_action='ignore')
```

```python
serie2 = pd.Series(np.arange(10))
serie2
```

```python
serie2.map(lambda a : a + 10)
```

#Apply

Pandas.apply permite aos usuários passar uma função e aplicá-la em cada valor da série Pandas. É uma grande melhoria para a biblioteca do pandas, uma vez que esta função ajuda a separar os dados de acordo com as condições exigidas, devido ao qual é usado de forma eficiente em ciência de dados e machine learning.

```python
df = pd.read_csv('sample_data/movies.csv', delimiter=',')

df.head(10)
```

```python
def converte_em_reais(valor):
  return 'R$ '+str(round(float(valor.replace('$',''))*5.09,2))
```

```python
df['Worldwide Gross'] = df['Worldwide Gross'].apply(converte_em_reais)
df.head(10)
```

#Applymap

Python é uma ótima linguagem para fazer análise de dados, principalmente por causa do fantástico ecossistema de pacotes python centrados em dados. O Pandas é um desses pacotes e torna a importação e análise de dados muito mais fácil.

Dataframe.applymap() método aplica uma função que aceita e retorna um escalar para cada elemento de um DataFrame.

```python
df = pd.DataFrame({
    'Col 1': [30,40,50,60],
    'Col 2': [23,35,65,45],
    'Col 3': [85,87,90,89],

},index=["A","B","C","D"])
df
```

```python
new_df=df.applymap(lambda a: a*10)
new_df
```

#Estudo de Caso 2:
Como alterar valores de uma tabela a partir de regras específicas?

```python
df = pd.read_csv('sample_data/ufo.csv', delimiter=',')
df.head(10)
```

```python
df.replace('NY','MA').head(20)
```

```python
def troca(estado):
  if estado=='CA':
    return 'SP'
  elif estado == 'NY':
    return 'MA'
  else:
    return estado
```

```python
df['State'] = df['State'].apply(troca)
df.head(20)
```
