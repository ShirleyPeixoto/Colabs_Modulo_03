#Filtrando um Data Frame

```python
#Vamos montar uma lista de booleanos que contemple nossa condição
booleanos = []

for i in df['Rotten Tomatoes %']:
  if (i > 60):
    booleanos.append(True)
  else:
    booleanos.append(False)

booleanos
```

```python
#verificando o tipo
type(booleanos)
```

```python
#comparando tamanho
len(booleanos)
```

```python
#comparando tamanho
df.shape
```

#Múltiplos Filtros

```python
#Filtro E
filtro1 = df['Rotten Tomatoes %'] >= 40
filtro2 = df['Rotten Tomatoes %'] <= 60
filmes_medianos = df[filtro1 & filtro2]
filmes_medianos
```

```python
#Filtro E
filmes_medianos = df[(df['Rotten Tomatoes %'] >= 40) & (df['Rotten Tomatoes %'] <= 60)]#Múltiplos Filtros
filmes_medianos
```

```python
#Filtro OU
filtro1 = df['Genre'] == 'Comedy'
filtro2 = df['Genre'] == 'Drama'
filtro3 = df['Genre'] == 'Animation'
filmes_comedia_ou_drama = df[filtro1 | filtro2 | filtro3] #Filtro OU
filmes_comedia_ou_drama
```

```python
#Filtro OU
filmes_comedia_ou_drama = df[(df['Genre'] == 'Comedy') | (df['Genre'] == 'Drama')]
filmes_comedia_ou_drama.head()
```
